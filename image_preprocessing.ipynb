```python
import cv2
os.chdir('images_png/무_train_labeled_2000')
files = os.listdir()
for file in files:
    # 이미지 파일을 읽습니다.
    image = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    original = cv2.imread(file)
    
    # 이진화
    _, binary = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY)

    # 컨투어 찾기
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    max_area = 0
    best_cnt = None
    # 눈알 영역만 추출
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > max_area:
            max_area = area
            best_cnt = contour
    
    x, y, w, h = cv2.boundingRect(best_cnt)
    cropped_img = original[y:y+h, x:x+w]

    # 사이즈 224x224로 변경
    resize_eye = cv2.resize(cropped_img, dsize=(224, 224), interpolation=cv2.INTER_AREA)
    cv2.imwrite(file, resize_eye)
```
